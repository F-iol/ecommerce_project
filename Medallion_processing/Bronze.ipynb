```python
from pyspark.sql import functions as F
```

## Brand

```python
_brand_schema = 'brand_code string NOT NULL,brand_name string,category_code string'
```

```python
brand_data='/Volumes/ecommerce/data_source/raw/brands/brands.csv'
```

```python
brand_df  = spark.read.option('header',True).schema(_brand_schema).option('delimeter',',').csv(brand_data)
```

```python
display(brand_df.show(3))
```

```python
brand_df = brand_df.withColumn('_source_file',F.col('_metadata.file_path')).withColumn('ingested_at',F.current_timestamp())
```

```python
brand_df.show(3,truncate=False)
```

```python
brand_df.write.format('delta').mode('overwrite').option('mergeSchema',True).saveAsTable('ecommerce.bronze.bronze_brands')
```

## Order Items

```python
_order_schema = 'dt string,order_ts string,customer_id string,order_id int,item_seq int,product_id string,quantity string,unit_price_currency string,unit_price string,discount_pct string,tax_amount int,channel string,coupon_code string'
_order_data= '/Volumes/ecommerce/data_source/raw/order_items/landing/*.csv'
```

```python
order_df = spark.read.option('header',True).schema(_order_schema).option('delimeter',',').csv(_order_data)
display(order_df.limit(1))
```

```python
order_df = order_df.withColumn('ingested_at',F.current_timestamp()).withColumn('_source_file',F.col('_metadata.file_path'))
display(order_df.limit(1))
```

```python
order_df.write.format('delta').mode('overwrite').option('overwriteSchema',True).saveAsTable('ecommerce.bronze.bronze_orders')
```
